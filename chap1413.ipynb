#### Hough Tranform in OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('hmson1.avi_000079750.png',0)
img2 = img.copy()
template = cv2.imread('20201105_125557.png',0)
```

```python
img = cv2.imread('sIMG_2164.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
lines = cv2.HoughLines(edges,1,np.pi/180,300)
```

```python
lines.shape
```

```python


for rho,theta in lines[0]:
 a = np.cos(theta)
 b = np.sin(theta)
 x0 = a*rho
 y0 = b*rho
 x1 = int(x0 + 1000*(-b))
 y1 = int(y0 + 1000*(a))
 x2 = int(x0- 1000*(-b))
 y2 = int(y0- 1000*(a))
 cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
cv2.imwrite('houghlines3.jpg',img)
```

```python
plt.imshow(img)
```

```python
for i in range(500):
 for rho,theta in lines[i]:
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0- 1000*(-b))
    y2 = int(y0- 1000*(a))
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
cv2.imwrite('houghlines3.jpg',img)
```

```python
def nothing(): 
    pass
```

```python
img = cv2.imread('sIMG_2164.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
minLineLength = 100
maxLineGap = 10
LineNums = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
```

```python

cv2.namedWindow('image') #imgae라는 이름의 윈도우 생성
# create trackbars for color change
# R, G, B라는 이름의 트랙바 생성, 트랙바 값이 초기값 0, 최대값 255, 트랙바 값 변경될 때 함수 호출
cv2.createTrackbar('minLineLength','image',0,500,nothing) 
#'R', 'G', 'B' 트랙바를 생성하여 0에서 255 사이의 값을 설정
cv2.createTrackbar('maxLineGap','image',0,500,nothing)
cv2.createTrackbar('LineNums','image',0,100,nothing)
cv2.setTrackbarPos('minLineLength','image',minLineLength)
cv2.setTrackbarPos('maxLineGap','image',maxLineGap)
cv2.setTrackbarPos('LineNums','image',LineNums)
# create switch for ON/OFF functionality

while(1):
    cv2.imshow('image',img) #img를 입력받아 'image'창이름으로 이미지를 윈도우에 표시
    k = cv2.waitKey(1) & 0xFF #1밀리초동안 대기하고 ESC 누를시 종료
    if k == 27:
        break
    # get current positions of four trackbars
    minLineLength = cv2.getTrackbarPos('minLineLength','image') #'R'이라는 트랙바에서 현재값을 가져와 r에 저장
    maxLineGap = cv2.getTrackbarPos('maxLineGap','image')
    LineNums = cv2.getTrackbarPos("LineNums",'image')
    lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
    LineNums = min(LineNums, lines.shape[0])
    for i in range(LineNums):
      for x1,y1,x2,y2 in lines[i]:
       cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
      
    #현재 switch 트랙바의 현재위치값(트랙바의 상태)을 가져와 s에 저장
    
cv2.destroyAllWindows()
```
